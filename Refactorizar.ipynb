```
<!DOCTYPE html>
<html lang="es" class="antialiased">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Nudo Gordiano Económico | Paper Ejecutable</title>

    <script src="https://cdn.tailwindcss.com"></script>
    <link href="https://fonts.googleapis.com/css2?family=Fira+Code:wght@400;500&family=Inter:wght@300;400;500;600&family=Merriweather:ital,wght@0,300;0,400;0,700;1,300&display=swap" rel="stylesheet">

    <script>
        window.MathJax = {
            tex: { inlineMath: [['$', '$'], ['\\(', '\\)']], displayMath: [['$$', '$$'], ['\\[', '\\]']] },
            chtml: { scale: 0.95 },
            startup: {
                pageReady: () => {
                    return MathJax.startup.defaultPageReady().then(() => {
                        document.getElementById('math-status').innerText = "MathJax: Activo";
                        document.getElementById('math-status').classList.replace('text-amber-500', 'text-teal-500');
                    });
                }
            }
        };
    </script>
    <script id="MathJax-script" async src="https://cdn.jsdelivr.net/npm/mathjax@3/es5/tex-chtml.js"></script>

    <style>
        :root {
            --bg-base: #0b1120; /* Slate 950 */
            --bg-panel: #1e293b; /* Slate 800 */
            --border-subtle: #334155;
            --text-main: #e2e8f0;
            --text-muted: #94a3b8;
            --text-prose: #cbd5e1;
            --accent: #14b8a6; /* Teal 500 */
        }
        body { background-color: var(--bg-base); color: var(--text-main); font-family: 'Inter', sans-serif; }

        /* Tipografía para lectura densa académica */
        .prose-academic {
            font-family: 'Merriweather', serif;
            line-height: 1.8;
            color: var(--text-prose);
            font-size: 1.05rem;
        }
        .prose-academic h2 { font-family: 'Inter', sans-serif; color: #f8fafc; font-weight: 600; margin-top: 2.5rem; margin-bottom: 1rem; font-size: 1.5rem; border-bottom: 1px solid var(--border-subtle); padding-bottom: 0.5rem; }
        .prose-academic h3 { font-family: 'Inter', sans-serif; color: #e2e8f0; font-weight: 500; margin-top: 2rem; margin-bottom: 0.75rem; font-size: 1.25rem; }
        .prose-academic p { margin-bottom: 1.25rem; text-align: justify; }
        .prose-academic strong { color: #f1f5f9; font-weight: 700; }

        .data-mono { font-family: 'Fira Code', monospace; font-variant-numeric: tabular-nums; }

        .panel {
            background: var(--bg-panel); border: 1px solid var(--border-subtle);
            border-radius: 4px; padding: 1.5rem; transition: border-color 0.2s;
        }

        .chart-container {
            position: relative; height: 500px; width: 100%;
            background: var(--bg-base); border: 1px solid var(--border-subtle); border-radius: 4px;
            background-image: linear-gradient(rgba(51, 65, 85, 0.4) 1px, transparent 1px),
                              linear-gradient(90deg, rgba(51, 65, 85, 0.4) 1px, transparent 1px);
            background-size: 50px 50px; background-position: center bottom;
        }
        canvas { display: block; width: 100%; height: 100%; }

        /* Custom Range Sliders */
        input[type=range] { -webkit-appearance: none; width: 100%; background: transparent; }
        input[type=range]::-webkit-slider-thumb { -webkit-appearance: none; height: 14px; width: 14px; border-radius: 50%; background: var(--accent); cursor: pointer; margin-top: -6px; }
        input[type=range]::-webkit-slider-runnable-track { width: 100%; height: 2px; cursor: pointer; background: var(--border-subtle); }
    </style>
</head>
<body class="min-h-screen p-4 md:p-8 lg:p-12 selection:bg-teal-500/30">

    <main class="max-w-4xl mx-auto space-y-12">

        <!-- ENCABEZADO Y METADATA -->
        <header class="border-b border-slate-700 pb-8">
            <h1 class="text-3xl md:text-5xl font-semibold tracking-tight text-white mb-4 leading-tight">El Nudo Gordiano de la Economía Chilena</h1>
            <p class="text-slate-400 font-medium text-lg md:text-xl mb-6">Modelado Estocástico de Ineficiencias Estructurales y Pérdida de Bienestar Agregado</p>

            <div class="flex flex-col md:flex-row justify-between items-start md:items-center bg-slate-800/50 p-4 border border-slate-700 rounded data-mono text-xs text-slate-400">
                <div class="flex flex-col gap-1">
                    <span>AUTOR: Hans Soriano</span>
                    <span>VERSIÓN: <span class="text-teal-500">v3.0.0-integral</span></span>
                </div>
                <div class="mt-2 md:mt-0 text-right">
                    <span id="math-status" class="text-amber-500">MathJax: Inicializando...</span>
                </div>
            </div>
        </header>

        <!-- CONTENIDO NARRATIVO ACADÉMICO -->
        <article class="prose-academic">

            <div class="border-l-4 border-teal-500 bg-teal-500/5 p-6 rounded-r mb-10">
                <p class="text-sm md:text-base mb-0 italic">
                    <strong>Abstract.</strong> Este documento ejecutable postula y cuantifica la histéresis del desarrollo chileno mediante un marco de cuatro vectores estocásticos acoplados. Se abandona la retórica política para modelar matemáticamente la degradación del ingreso ciudadano neto ($I_{neto}$) como función de la hipertrofia de agencia, la búsqueda de rentas (rent-seeking) inmobiliaria, la indexación regresiva de la matriz logística y la varianza exógena. El colapso de la Productividad Total de Factores (PTF) se evidencia aquí no como un accidente, sino como el resultado determinista de una arquitectura institucional ineficiente.
                </p>
            </div>

            <h2>I. Formulación del Sistema y Función de Transferencia</h2>
            <p>
                El paradigma económico local ha sostenido históricamente que el estancamiento es producto exclusivo de *shocks* externos o de ciclos crediticios globales. Sin embargo, al modelar la estructura de costos a nivel micro, se observa que el ingreso ciudadano está sujeto a un vector de fricciones institucionales endógenas.
            </p>
            <p>
                Definimos la ecuación fundamental de transferencia de bienestar, la cual determina la esperanza matemática del ingreso en el periodo $t$, como un sistema de sustracción sobre un rendimiento teórico base ($Y_0$):
            </p>

            <div class="bg-slate-900 p-6 rounded border border-slate-700 my-6 overflow-x-auto text-sm md:text-base">
                $$I_{neto,t} = \underbrace{Y_0 \cdot \mathcal{E}(k)}_{\text{Base ajustada}} - \underbrace{\delta_E(\beta+\tau)}_{\text{Fricción Fija}} - \underbrace{\delta_C\mu_{m^2}^{eff}}_{\text{Exacción Inmob.}} - \underbrace{\phi B_t}_{\text{Shock}} + \eta_t$$
            </div>

            <p>
                Este modelo reconoce que las fallas de mercado y de gobierno no son aditivas, sino que se amplifican mutuamente. La varianza del sistema ($\text{Var}[\eta_t]$), que captura el ruido blanco propagado desde la volatilidad de precios hasta el consumo final, se define como:
            </p>

            <div class="bg-slate-900 p-6 rounded border border-slate-700 my-6 overflow-x-auto text-sm md:text-base">
                $$\text{Var}[\eta_t] = \delta_E^2\sigma_\varepsilon^2 + \delta_C^2\sigma_{m^2}^2 = 903.24 \text{ CLP}^2$$
            </div>

            <h2>II. Topología del Estancamiento: Los Cuatro Pilares</h2>

            <h3>1. Costos de Agencia e Hipertrofia Burocrática ($k$)</h3>
            <p>
                El primer sumidero de eficiencia corresponde a la estructura organizativa del Estado. Modelamos el costo de agencia mediante una función de decrecimiento $\mathcal{E}(k) = \max(0.05, 1 - 0.09k)$, donde $k$ representa las capas de administración superpuestas. La arquitectura actual (GORE, Delegados Presidenciales, SEREMIs, Municipios, etc.) configura una línea base subóptima ($k \ge 5$), destruyendo un porcentaje masivo del rendimiento productivo antes de que el capital se despliegue en la economía real.
            </p>

            <h3>2. Búsqueda de Rentas y Exacción Inmobiliaria ($s$)</h3>
            <p>
                El mercado de suelo urbano en Chile exhibe profundas ineficiencias derivadas de la captura regulatoria (planes reguladores restrictivos) y la especulación financiera. La desviación especulativa del costo por metro cuadrado se modela como $\mu_{m^2}^{eff} = \mu_{m^2}(1+s)$. Este sobreprecio ($s$) opera como un impuesto privado, drenando liquidez del sector productivo hacia la acumulación pasiva de capital y castigando directamente el ingreso disponible de los hogares.
            </p>

            <h3>3. Indexación Regresiva de la Matriz Energética ($\tau$)</h3>
            <p>
                La economía nacional posee una dependencia inelástica del transporte terrestre. El gravamen específico a los combustibles ($\tau$) opera como una cuña fiscal profundamente regresiva. Dado que el costo de la logística se transfiere en su totalidad al precio de la canasta básica a través del coeficiente de sensibilidad $\delta_E$, el diseño tributario actual amplifica la inflación de costos, mermando el salario real de forma silenciosa pero sistémica.
            </p>

            <h3>4. Commoditización y Varianza Exógena ($B_t$)</h3>
            <p>
                La matriz productiva, al carecer de complejidad económica y valor agregado tecnológico, queda expuesta de manera desnuda a las fluctuaciones internacionales de materias primas. Modelamos el precio del crudo Brent ($B_t$) mediante un proceso estocástico (Random Walk con reversión a la media). La incapacidad de la industria local para absorber estos *shocks* se traduce en una transferencia directa de la volatilidad global hacia el bolsillo del ciudadano promedio.
            </p>

        </article>

        <!-- SECCIÓN INTERACTIVA / SIMULADOR -->
        <section class="border-t border-slate-700 pt-12">
            <div class="mb-8">
                <h2 class="text-2xl font-semibold text-white mb-2">III. Laboratorio Estocástico y Simulación Dinámica</h2>
                <p class="text-slate-400 text-sm">Altera las fricciones estructurales en tiempo real para observar el impacto en la esperanza matemática del ingreso ($\mathbb{E}[I_{neto}]$) y la convergencia de las iteraciones de Montecarlo.</p>
            </div>

            <div class="flex flex-col lg:flex-row gap-8">
                <!-- Controles -->
                <div class="w-full lg:w-1/3 flex flex-col gap-6">

                    <div class="bg-slate-900/80 p-5 rounded border border-slate-700">
                        <div class="text-xs font-semibold text-slate-400 uppercase tracking-wider mb-4 border-b border-slate-700 pb-2">Esperanza Analítica</div>
                        <div class="data-mono text-3xl text-teal-400 font-bold mb-1">
                            $<span id="e-neto-display">...</span>
                        </div>
                        <div class="text-xs text-slate-500 font-mono">CLP / Rendimiento proyectado</div>
                    </div>

                    <div class="panel p-5">
                        <label class="flex justify-between text-sm font-semibold text-slate-200 mb-3">
                            <span>Capas Burocráticas ($k$)</span>
                            <span id="val-k" class="data-mono text-rose-400">5</span>
                        </label>
                        <input type="range" id="input-k" min="1" max="10" step="1" value="5">
                        <p class="text-xs text-slate-500 mt-3 leading-relaxed">Determina el peso del Estado. Valores $\ge 5$ reflejan ineficiencia administrativa aguda.</p>
                    </div>

                    <div class="panel p-5">
                        <label class="flex justify-between text-sm font-semibold text-slate-200 mb-3">
                            <span>Especulación de Suelo ($s$)</span>
                            <span id="val-s" class="data-mono text-amber-400">0%</span>
                        </label>
                        <input type="range" id="input-s" min="0" max="100" step="5" value="0">
                        <p class="text-xs text-slate-500 mt-3 leading-relaxed">Aumento porcentual artificial del valor del m² producto del *rent-seeking* financiero.</p>
                    </div>

                    <div class="panel p-5">
                        <label class="flex justify-between text-sm font-semibold text-slate-200 mb-3">
                            <span>Impuesto Logístico ($\tau$)</span>
                            <span id="val-tau" class="data-mono text-teal-400">222</span>
                        </label>
                        <input type="range" id="input-tau" min="0" max="500" step="10" value="222">
                        <p class="text-xs text-slate-500 mt-3 leading-relaxed">Impuesto específico base (CLP/litro). Encarece de manera transversal toda la cadena de suministro.</p>
                    </div>
                </div>

                <!-- Canvas -->
                <div class="w-full lg:w-2/3">
                    <div class="chart-container shadow-inner">
                        <canvas id="incomeChart"></canvas>
                    </div>
                    <p class="text-center text-[11px] text-slate-500 mt-4 font-mono leading-relaxed max-w-lg mx-auto">
                        Fig 1. Trayectorias simuladas del Ingreso Neto ciudadano a $t=120$ meses mediante caminata aleatoria. Banda sombreada indica intervalo de confianza ($\mu \pm 3\sigma$).
                    </p>
                </div>
            </div>
        </section>

    </main>

    <!-- MOTOR DE SIMULACIÓN -->
    <script>
        const Y0 = 1182000;
        const DELTA_C = 0.01;
        const DELTA_E = 0.12;
        const MU_M2_BASE = 30000;
        const BETA = 550.93;
        const ALPHA = 8.4486;
        const BRENT_MU = 75;
        const VAR_ETA = 903.24;
        const SIGMA_TOTAL = Math.sqrt(VAR_ETA);
        const N_MONTHS = 120;
        const N_PATHS = 5;

        let E_I_NETO = 0;
        let pathsData = [];

        const inputK = document.getElementById('input-k');
        const inputS = document.getElementById('input-s');
        const inputTau = document.getElementById('input-tau');
        const valK = document.getElementById('val-k');
        const valS = document.getElementById('val-s');
        const valTau = document.getElementById('val-tau');
        const displayENeto = document.getElementById('e-neto-display');

        const canvas = document.getElementById('incomeChart');
        const ctx = canvas.getContext('2d');
        let dpr = window.devicePixelRatio || 1;

        function updateMath() {
            const k = parseFloat(inputK.value);
            const s = parseFloat(inputS.value) / 100;
            const tau = parseFloat(inputTau.value);

            valK.innerText = k;
            valS.innerText = `${(s * 100).toFixed(0)}%`;
            valTau.innerText = tau;

            const E_k = Math.max(0.05, 1 - 0.09 * k);
            const mu_eff = MU_M2_BASE * (1 + s);
            const energy_friction = DELTA_E * (ALPHA * BRENT_MU + BETA + tau);

            E_I_NETO = (Y0 * E_k) - energy_friction - (DELTA_C * mu_eff);
            displayENeto.innerText = Math.round(E_I_NETO).toLocaleString('es-CL');
        }

        function generateData() {
            pathsData = [];
            for(let p=0; p<N_PATHS; p++) {
                let currentPath = [];
                let current_val = E_I_NETO;
                for(let i=0; i<N_MONTHS; i++) {
                    let u = 0, v = 0;
                    while(u === 0) u = Math.random();
                    while(v === 0) v = Math.random();
                    let z = Math.sqrt(-2.0 * Math.log(u)) * Math.cos(2.0 * Math.PI * v);

                    let noise = z * SIGMA_TOTAL * 15;
                    current_val = current_val * 0.9 + (E_I_NETO + noise) * 0.1;
                    currentPath.push(current_val);
                }
                pathsData.push(currentPath);
            }
        }

        function resizeCanvas() {
            const rect = canvas.parentNode.getBoundingClientRect();
            canvas.width = rect.width * dpr;
            canvas.height = rect.height * dpr;
            ctx.scale(dpr, dpr);
        }

        function drawChart() {
            if (!pathsData.length) return;
            const w = canvas.width / dpr;
            const h = canvas.height / dpr;
            ctx.clearRect(0, 0, w, h);

            const p = { t: 40, b: 40, l: 80, r: 20 };
            const cw = w - p.l - p.r;
            const ch = h - p.t - p.b;

            const span = 150000;
            const minV = E_I_NETO - span;
            const maxV = E_I_NETO + span;
            const rng = maxV - minV;

            const getY = val => p.t + ch - ((val - minV) / rng) * ch;
            const getX = idx => p.l + (idx / (N_MONTHS - 1)) * cw;

            const sigmaViz = SIGMA_TOTAL * 15;
            ctx.fillStyle = 'rgba(20, 184, 166, 0.05)';
            ctx.fillRect(p.l, getY(E_I_NETO + sigmaViz*3), cw, getY(E_I_NETO - sigmaViz*3) - getY(E_I_NETO + sigmaViz*3));

            ctx.strokeStyle = '#334155'; ctx.lineWidth = 1; ctx.beginPath();
            ctx.moveTo(p.l, p.t); ctx.lineTo(p.l, p.t + ch); ctx.lineTo(p.l + cw, p.t + ch); ctx.stroke();

            ctx.strokeStyle = '#f59e0b'; ctx.setLineDash([4, 4]); ctx.beginPath();
            ctx.moveTo(p.l, getY(E_I_NETO)); ctx.lineTo(p.l + cw, getY(E_I_NETO)); ctx.stroke(); ctx.setLineDash([]);

            ctx.fillStyle = '#94a3b8'; ctx.font = '11px "Fira Code"'; ctx.textAlign = 'right'; ctx.textBaseline = 'middle';
            ctx.fillText(Math.round(maxV).toLocaleString('es-CL'), p.l - 10, p.t);
            ctx.fillText(Math.round(minV).toLocaleString('es-CL'), p.l - 10, p.t + ch);
            ctx.fillStyle = '#f59e0b';
            ctx.fillText(`\u00b5 = ${Math.round(E_I_NETO).toLocaleString('es-CL')}`, p.l - 10, getY(E_I_NETO));

            const colors = ['#14b8a6', '#0ea5e9', '#8b5cf6', '#f43f5e', '#64748b'];
            pathsData.forEach((path, idx) => {
                ctx.strokeStyle = colors[idx % colors.length];
                ctx.globalAlpha = 0.6;
                ctx.lineWidth = 1.5;
                ctx.lineJoin = 'round';
                ctx.beginPath();
                ctx.moveTo(getX(0), getY(path[0]));
                for(let i=1; i<N_MONTHS; i++) ctx.lineTo(getX(i), getY(path[i]));
                ctx.stroke();
            });
            ctx.globalAlpha = 1.0;
        }

        function runSimulation() {
            updateMath();
            generateData();
            drawChart();
        }

        window.addEventListener('resize', () => { resizeCanvas(); drawChart(); });
        [inputK, inputS, inputTau].forEach(el => el.addEventListener('input', runSimulation));

        resizeCanvas();
        runSimulation();
    </script>
</body>
</html>
```
